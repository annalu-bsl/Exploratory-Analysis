```r
data <- read.csv("ava_manu_ad.csv", sep = ';', header = )
```

```r
data
length(data$Processo.MA)
```

```r
library(ggplot2)
library(dplyr)
```

```r
table(data$v..mm.s.)
data <- data[,-15]
data <- data[,-14]
data <- data[,-13]
data <- data[,-12]
data <- data[,-11]
data <- data[,-10]
data <- data[,-9]
data
```

```r
colnames(data) <- c("processo", "alloy", "pot", "velocidade", "H_mm", "ref", "metal", "Tf")
data
```

```r
newdata <- with(data,data[!(data$velocidade == ""),])
newdata <- newdata[-101,]
newdata <- newdata[-100,]
newdata$velocidade <- as.double(newdata$velocidade)
```

```r
newdata
data1 <- with(newdata, newdata[!(newdata$velocidade == 10000),])
data1
```

```r
plotvel <- ggplot(newdata, aes(x = velocidade, y = Tf)) + geom_point() + ggtitle("TF x Velocidade")
plotvel
plotvel1 <- ggplot(data1, aes(x = velocidade, y = Tf)) + geom_point() + ggtitle("TF x Velocidade sem Outliers")
plotvel1
bp1 <- ggplot(data = newdata, aes(x = Tf, y = velocidade, fill = Tf, group = Tf)) + geom_boxplot() + ggtitle("Boxplot Velocidade x TF")
bp1
boxp <- ggplot(data = data1, aes(x = Tf, y = velocidade, fill = Tf, group = Tf)) + geom_boxplot() + ggtitle("Boxplot Velocidade x TF sem Outliers")
boxp
histtf <- hist(newdata$Tf, xlab = "TF", main = "Histograma TF", col = c("light blue"))
histtf
table(newdata$velocidade)
histvel <- hist(newdata$velocidade, xlab = "velocidade", main = "Histograma Velocidades", col = c("light blue"))
histvel
```

```r
plotpot <- ggplot(newdata, aes(x = pot, y = Tf)) + geom_point() +ggtitle("TF x Potência")
plotpot
bp2 <- ggplot(data = newdata, aes(x = Tf, y = pot, fill = Tf, group = Tf)) + geom_boxplot()+ggtitle("Boxplot Potência x TF")
bp2
histpot <- hist(newdata$pot, xlab = "potência", main = "Histograma Potências", col = c("light blue"))
histpot
```

```r
data2 <- with(newdata, newdata[!(newdata$pot >1500),])
data2
```

```r
boxp2 <- ggplot(data = data2, aes(x = Tf, y = pot, fill = Tf, group = Tf)) + geom_boxplot() + ggtitle("Boxplot Potência x TF")
boxp2
max(newdata$pot, na.rm = TRUE) #temperatura mais alta
```

```r
plotHmm <- ggplot(newdata, aes(x = H_mm, y = Tf)) + geom_point() + ggtitle("TF x H/mm") 
plotHmm
bp3 <- ggplot(data = newdata, aes(x = Tf, y = H_mm, fill = Tf, group = Tf)) + geom_boxplot()+ ggtitle("Boxplot H/mm x TF") 
bp3
table(newdata$H_mm)
histHmm <- hist(newdata$H_mm, xlab = "H/mm", main = "Histograma H/mm", col = c("light blue"))
histHmm
```

```r
data3 <- with(newdata, newdata[!(newdata$H_mm >1),])
boxp3 <- ggplot(data = data3, aes(x = Tf, y = H_mm, fill = Tf, group = Tf)) + geom_boxplot() + ggtitle("Boxplot H/mm x TF sem Outliers")
boxp3
```

```r
plotvelpot <- ggplot(newdata, aes(x = velocidade, y = pot, colour = Tf)) + geom_point() + ggtitle("Potência x Velocidade")
plotvelpot
plotvelpot2<- ggplot(data2, aes(x = velocidade, y = pot, colour = Tf)) + geom_point()+ ggtitle("Potência x Velocidade sem Outliers")
plotvelpot2
dataplot <- with(data2, data2[!(data2$velocidade > 3000),])
plotvelpotprev<- ggplot(dataplot, aes(x = velocidade, y = pot, colour = Tf)) + geom_point() + geom_smooth() + ggtitle("Previsão de Potência x Velocidade Sem Outliers")
plotvelpotprev
ggplot(dataplot, aes(x = velocidade, y = pot)) + geom_point() + geom_smooth(method = "lm")
```

```r
plotvelHmm <- ggplot(newdata, aes(x = velocidade, y = H_mm, colour = Tf)) + geom_point() + ggtitle("Velocidade x H/mm")
plotvelHmm
ggplot(data3, aes(x = velocidade, y = H_mm, colour = Tf)) + geom_point() + ggtitle("Velocidade x H/mm sem Outliers")
dataplot2 <- with(data3, data3[!(data3$velocidade > 3000),])
plotvelHmm2 <- ggplot(dataplot2, aes(x = velocidade, y = H_mm, colour = Tf)) + geom_point() + ggtitle("Velocidade x H/mm Sem Outliers")
plotvelHmm2 
```

```r
plotpotHmm <- ggplot(newdata, aes(x = pot, y = H_mm, colour = Tf)) + geom_point() + ggtitle("H/mm x Potência")
plotpotHmm 
ggplot(data3, aes(x = pot, y = H_mm, colour = Tf)) + geom_point() + ggtitle("H/mm x Potência")
dataplot3 <- with(data3, data3[!(data3$pot > 1000),])
plotpotHmm2 <- ggplot(dataplot3, aes(x = velocidade, y = H_mm, colour = Tf)) + geom_point()+ ggtitle("H/mm x Potência Sem Outliers")
plotpotHmm2
```
